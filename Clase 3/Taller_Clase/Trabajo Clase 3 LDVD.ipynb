## Negativo de una imagen

Este punto consiste en tomar la imagen "bacteria.jpeg" y convertirla a su negativo. El negativo de una imagen se obtiene restando a cada pixel el valor 255, pero aquí está la novedad: vamos a realizar el negativo no en escala de grises, sino en cada canal de color por separado. Es decir, si un pixel tiene el valor (r, g, b), el negativo de ese pixel será (255-r, 255-g, 255-b).

### Imagen original

```python
Image("res/Actividad1/bacteria.jpeg")
```

### Imagen deseada

```python
Image("res/Actividad1/negativo.png")
```

```python
# Solución Actividad 1
img = cv2.imread("res/Actividad1/bacteria.jpeg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

colors = ('r','g','b')
img_neg = np.zeros(img.shape, dtype=np.uint8)
for i in range(3):
    img_neg[:,:,i] = 255 - img[:,:,i]

fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(15, 15))
ax1.imshow(img)
ax1.set_title("Imagen original")
for i, col in enumerate(colors):
    img_array_i = img[:,:,i].ravel()
    ax2.hist(img_array_i, histtype='step', bins=255, range=(0.0, 255.0), density=True, color=colors[i])
ax3.imshow(img_neg)
ax3.set_title("Imagen negativa")
for i, col in enumerate(colors):
    img_array_i = img_neg[:,:,i].ravel()
    ax4.hist(img_array_i, histtype='step', bins=255, range=(0.0, 255.0), density=True, color=colors[i])
```

## Encoger el histograma

Disminuya el histograma de la imagen "teatro.jpg" en un 50%. El resultado esperado son cuatro imágenes: la imagen original, el histograma original, la imagen modificada y el histograma modificado.

```python
Image("res/Actividad2/teatro.jpg")
```

```python
# Solución Actividad 2
colors = ('r','g','b')

teatro = cv2.imread("res/Actividad2/teatro.jpg")
teatro = cv2.cvtColor(teatro, cv2.COLOR_BGR2RGB)
teatroMod = np.zeros(teatro.shape, dtype=np.uint8)
teatroMod[:,:] = teatro[:,:] * 0.5
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(15, 15))
ax1.imshow(teatro)
ax1.set_title("Imagen original")
for i, col in enumerate(colors):
    img_array_i = teatro[:,:,i].ravel()
    ax2.hist(img_array_i, histtype='step', bins=255, range=(0,255), color=colors[i])
ax3.imshow(teatroMod)
ax3.set_title("Imagen modificada")
for i, col in enumerate(colors):
    img_array_i = teatroMod[:,:,i].ravel()
    ax4.hist(img_array_i, histtype='step', bins=255, range=(0,255), density = True, color=colors[i])
```
