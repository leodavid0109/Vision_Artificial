# Sheet Code

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

## Transformaciones geom√©tricas

```python
doge = cv2.imread('res/doge.jpg')
doge = cv2.cvtColor(doge, cv2.COLOR_BGR2RGB)
plt.imshow(doge)
```

### Traslaci√≥n

```python
tx = 20
ty = 20
matriz_afin = np.float32([[1,0,tx],
                          [0,1,ty]])
dogeTraslacion = cv2.warpAffine(doge, matriz_afin, (doge.shape[1], doge.shape[0]))
plt.imshow(dogeTraslacion)
```

# Reflexi√≥n

### X

```python
matriz_afin_ref_x = np.float32([[-1,0,doge.shape[1]]
                                ,[0,1,0]])
dogeReflexionX = cv2.warpAffine(doge, matriz_afin_ref_x, (doge.shape[1], doge.shape[0]))
plt.imshow(dogeReflexionX)
```

### Y

```python
matriz_afin_ref_y = np.float32([[1,0,0],
                                [0,-1,doge.shape[0]]])
dogeReflexionY = cv2.warpAffine(doge, matriz_afin_ref_y, (doge.shape[1], doge.shape[0]))
plt.imshow(dogeReflexionY)
```

## Rotaci√≥n

```python
theta = 90
rotation_center = ((doge.shape[1]/2,doge.shape[0]/2))
matriz_afin_rot = cv2.getRotationMatrix2D(rotation_center,theta,1)
dogeRotacion = cv2.warpAffine(doge, matriz_afin_rot, (doge.shape[1], doge.shape[0]))
plt.imshow(dogeRotacion)
```

### Inclinaci√≥n

```python
inc_x = -0.4
inc_y = 0.0
matriz_afin = np.float32([[1,inc_x,0],
                          [inc_y,1,0]])
dogInclinado = cv2.warpAffine(doge, matriz_afin, (doge.shape[1], doge.shape[0]))
plt.imshow(dogInclinado)
```

### Escalamiento

```python
dogeEscaladoCercano = cv2.resize(doge, None, fx=1.5, fy=1.5, interpolation=cv2.INTER_NEAREST)
dogeEscaladoBilineal = cv2.resize(doge, None, fx=1.5, fy=1.5, interpolation=cv2.INTER_LINEAR)
dogeEscaladoCubico = cv2.resize(doge, None, fx=1.5, fy=1.5, interpolation=cv2.INTER_CUBIC)
_, plots = plt.subplots(1,3, figsize=(30,30))
plots[0].set_title('Vecino m√°s cercano')
plots[0].imshow(dogeEscaladoCercano)
plots[1].set_title('Bilineal')
plots[1].imshow(dogeEscaladoBilineal)
plots[2].set_title('C√∫bico')
plots[2].imshow(dogeEscaladoCubico)
```
